# Kva er poenget med denne modullen?

I denne modulen kjem vi til å studere frekvensar til forskjellige signaler. Signalene kan være til dømes straumsignal, lydsignal eller lyssignal. For å forstå hva vi gjør, og hva frekvenser er, kommer vi til å fokusere på lyd. Grunnen til dette er at ørene våre klarer å differensiere mellom forskjellige frekvenser. La oss begynne med å se på signal som består av ein enkel (reell) frekvens.

```python
# Importing pakages to do numerics and visualisation
import numpy as np
import matplotlib.pyplot as plt

# For showing and playing sound in Jupyter notebook.
import IPython
```

## Reine tonar og summen av dei

La oss sjå og høyre på nokre døme på kva som er meint med (reelle) frekvenser. For ein oversikt over kva frekvenser som korosponderar til kva for nokre tonar sjå her: https://pages.mtu.edu/~suits/notefreqs.html.

```python
# The sampling rate of the audio, measured in samples per second.
sampling_rate = 22050

# Length in secunds of the audio signal.
length_time = 5 
t = np.linspace(0, 3, sampling_rate * length_time, endpoint=False)
```

### C4

La oss ta for oss ein C tone med frekvens på 261,63Hz.

```python
frequency_c = 261.63 # Frequency of signal in Hz

# A pure middle C4 signal
middle_c = 0.5 * np.sin(2 * np.pi * frequency_c * t)
```

```python
plt.plot(t, middle_c)
plt.title("Middle C")
plt.xlabel("Time")
plt.ylabel("Preasure")
plt.xlim((0, 5* 1/262))

plt.show()
```

```python
# Run this cell and play by pressing the play button
IPython.display.Audio(middle_c, rate=sampling_rate)
```

### A4

```python
frequency_a4 = 440 # Frequency of signal in Hz

# A pure middle a4 signal
a4 = 0.3 * np.sin(2 * np.pi * frequency_a4 * t)
```

```python
plt.plot(t, a4)
plt.title("A4")
plt.xlabel("Time")
plt.ylabel("Preasure")
plt.xlim((0, 5* 1/262))

plt.show()
```

```python
# Run this cell and play by pressing the play button
IPython.display.Audio(a4, rate=sampling_rate)
```

### Summen av signala

```python
sum_c4_a4 = middle_c + a4
```

```python
plt.plot(t, sum_c4_a4)
# plt.plot(t, a4)
# plt.plot(t, middle_c)

plt.title("Sum of the tones C4 and A4")
plt.xlabel("Time")
plt.ylabel("Preasure")
plt.xlim((0, 5* 1/262))

plt.show()
```

```python
# Run this cell and play by pressing the play button
IPython.display.Audio(sum_c4_a4, rate=sampling_rate)
```
