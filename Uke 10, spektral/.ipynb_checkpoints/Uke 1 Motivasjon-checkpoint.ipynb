# Kva er poenget med denne modulen?

I denne modulen kjem vi til å studere frekvensar til forskjellige signaler. Signala kan være til dømes straumsignal, lydsignal eller lyssignal. For å forstå kva vi gjer, og kva frekvensar er, kjem vi til å fokusere på lyd. Grunnen til dette er at øyra våre klarer å differensiere mellom forskjellige frekvensar. La oss begynne med å sjå på signal som består av ein enkel (reell) frekvens.

```python
# Importing pakages to do numerics and visualisation
import numpy as np
import matplotlib.pyplot as plt

# For showing and playing sound in Jupyter notebook.
import IPython
```

## Reine tonar og summen av dei

La oss sjå og høyre på nokre døme på kva som er meint med (reelle) frekvensar. For ein oversikt over kva frekvensar som korresponderer til kva for nokre tonar sjå her: https://pages.mtu.edu/~suits/notefreqs.html.

```python
# The sampling rate of the audio, measured in samples per second.
sampling_rate = 22050

# Length in secunds of the audio signal.
length_time = 5 
t = np.linspace(0, 3, sampling_rate * length_time, endpoint=False)
```

### C4

La oss ta for oss ein C tone med frekvens på 261,63Hz.

```python
frequency_c = 261.63 # Frequency of signal in Hz

# A pure middle C4 signal
middle_c = 0.5 * np.sin(2 * np.pi * frequency_c * t)
```

```python
plt.plot(t, middle_c)
plt.title("C4 tone")
plt.xlabel("Tid")
plt.ylabel("Lufttrykk")
plt.xlim((0, 5* 1/262))

plt.show()
```

```python
# Run this cell and play by pressing the play button
IPython.display.Audio(middle_c, rate=sampling_rate)
```

### A4

```python
frequency_a4 = 440 # Frequency of signal in Hz

# A pure middle a4 signal
a4 = 0.3 * np.sin(2 * np.pi * frequency_a4 * t)
```

```python
plt.plot(t, a4)
plt.title("A4 tone")
plt.xlabel("Tid")
plt.ylabel("Lufttrykk")
plt.xlim((0, 5* 1/262))

plt.show()
```

```python
# Run this cell and play by pressing the play button
IPython.display.Audio(a4, rate=sampling_rate)
```

### Summen av signala

Når vi tek summen av to tonar, kan ein høyre dei meir eller mindre separert. Det vil sei at ein kan høyre kvar enkel tone i eit signal. For å bevise dette kan vi sjå at dette stemmer, kan vi ta summen av dei to signala over. Summen av dei to signala vil høyrast ut som om begge signala blir spelt av samtidig.

```python
# The sum of the two signals
sum_c4_a4 = middle_c + a4
```

```python
plt.plot(t, sum_c4_a4)
# plt.plot(t, a4)
# plt.plot(t, middle_c)

plt.title("Summen av C4 and A4")
plt.xlabel("Tid")
plt.ylabel("Lufttrykk")
plt.xlim((0, 5* 1/262))

plt.show()
```

```python
# Run this cell and play by pressing the play button
IPython.display.Audio(sum_c4_a4, rate=sampling_rate)
```
